```python
print(portfolioAnalysisRequest.text)
```

```python
import json
jsonRes=json.loads(portfolioAnalysisRequest.text)
print(jsonRes)
```

```python
for val in jsonRes:
    print(val,"\t",jsonRes[val]['level'])
print("\n\n")
```

```python
import csv
with open('values.csv', 'w') as csvfile:
    fieldnames = ['Date', 'Level']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for val in jsonRes:
        writer.writerow({'Date':val,'Level':jsonRes[val]['level']})
```
